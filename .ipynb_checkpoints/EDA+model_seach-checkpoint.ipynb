### Вес

Построим гистрограмму, ящик с усами, и исследуем описательную статистику.

```python
def hist_box(df: pd.DataFrame,
             col_name: str, 
             title: str='', 
             xlabel: str='',
             ylabel: str='',
             bins:int=20): 
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True,
                                        gridspec_kw={"height_ratios": (.15, .85)},
                                        figsize=(16, 8))
    sns.histplot(data=df, x=col_name, ax= ax_hist, bins=bins,  kde=True);
    sns.boxplot(data=df, x=col_name, ax= ax_box);
    ax_box.set(xlabel='')
    plt.title(title)
    ax_hist.set(xlabel=xlabel)
    ax_hist.set(ylabel=ylabel)
    plt.show()
```

```python
hist_box(df, "weight", "Распределение веса", "Вес, кг", "Количество")
```

## Анализ категориальной зависимой переменной

```python
df["size"].value_counts().plot(kind='pie', 
                               title="Размеры", 
                               fontsize=16, 
                               legend=False, 
                               autopct='%1.1f%%', 
                               figsize=(7,7));
```

```python
ax = df["size"].value_counts(ascending=True).plot(kind='barh', 
                               title="Размеры", 
                               fontsize=16,  
                               figsize=(10,7));
ax.bar_label(ax.containers[0]);
```

Имеется дисбаланс классов, что стоит учитыать при построении и оценци модели. Примеров XXL класса очень мало - имеет смысл от него отказаться. Данные принадлежат порядковой шкале. Закодируем.

```python
df = df[df['size'] != 'XXL']
```

```python
def encode_target(y: str) -> int:
    if y == 'XXS': return 1
    if y == 'S': return 2
    if y == 'M': return 3
    if y == 'L': return 4
    if y == 'XL': return 5
    if y == 'XXXL': return 6
    return -1
```

```python
df['size'] = df['size'].apply(encode_target)
df.head()
```

### Отличается ли средний возраст для двух популярных размеров

## Построение моделей

Выделим фичи и таргет.

```python
X = df.drop("size", axis=1)
y = df['size']
```

Выделим тренировочный и тестовый наборы. При дисбалансе классов разумнее использовать стратифицированную выборку.

```python
SEED = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED, stratify=y)
```

### Логистическая регрессия

Выделяем тренировочный и тестовый набор.

У логистической регрессии булем подбирать один гиперпараметр - С - степень регуляризации. Будем подбирать используя GridSearchCV.  Логистическая чувствительна к шкалированию, поэтому будем стандартизировать признаки, а чтобы они стандартизировались независимо при перекрёстной проверке будем использовать Pipeline. В качестве метрики из-за дисбаланса классов разумно использовать f1-меру.

```python
scaler = StandardScaler()
cv = StratifiedKFold(5)
model = LogisticRegression(random_state=SEED, class_weight="balanced", max_iter=5000)
pipe = Pipeline([("scaler", scaler), ("logreg", model)])
params = {"logreg__C": np.logspace(-4, 5, 10)}
lr = GridSearchCV(pipe, params, scoring='f1_weighted', cv=cv, verbose=3)
lr.fit(X_train, y_train)
```

```python
print(f'Лучшая модель имеет значение f1_weighted: {lr.best_score_:.3f} при C = {lr.best_params_["logreg__C"]:.3f}')
```

Сравним с dummy.

```python
dummy = DummyClassifier()
cross_val_score(dummy, X_train, y_train, scoring='f1_weighted', cv=cv).mean()
```

Лучше, чем дамми, но продолжим поиск.

### Метод ближайших соседей.

```python
model = KNeighborsClassifier(n_jobs=-1)
params = {"knn__n_neighbors":list(range(10, 100, 5))}
pipe = Pipeline([("scaler", scaler), ("knn", model)])
knn = GridSearchCV(pipe, params, scoring='f1_weighted', cv=cv, verbose=3)
knn.fit(X_train, y_train)
```

```python
print(f'Лучшая модель имеет значение f1_weighted: {knn.best_score_:.3f} при K = {knn.best_params_["knn__n_neighbors"]:d}')
```

### Дерево решений

```python
model = DecisionTreeClassifier(random_state=SEED, class_weight='balanced')
params = {'max_depth':range(2, 20)}
tree = GridSearchCV(model, params, scoring='f1_weighted', verbose=3, cv=cv)
tree.fit(X_train, y_train)
```

```python
print(f'Лучшая модель имеет значение f1_weighted: {tree.best_score_:.3f} при max_depth = {tree.best_params_["max_depth"]:d}')
```

### Случайный лес

```python
model = RandomForestClassifier(random_state=SEED, class_weight='balanced')
params = {'n_estimators':[50, 100, 200, 300, 400, 500], 
          'max_depth':range(10, 15)}
rf = GridSearchCV(model, params, scoring='f1_weighted', verbose=3, cv=cv)
rf.fit(X_train, y_train)
```

```python
print(f'Лучшая модель имеет значение f1_weighted: {rf.best_score_:.3f} при {rf.best_params_}')
```

```python
model = LGBMClassifier(random_state=SEED, class_weight='balanced')
params = {'learning_rate':[0.01, 0.1, 0.2, 0.3, 0.5, 0.75, 1.0], 
          'max_depth':range(20, 40, 5)
         }
gb = GridSearchCV(model, params, scoring='f1_weighted', verbose=3, cv=cv)
gb.fit(X_train, y_train)
```

```python
print(f'Лучшая модель имеет значение f1_weighted: {gb.best_score_:.3f} при {gb.best_params_}')
```

## Оценка лучшей модели
Лучшие результаты показали KNN и Случайный лес. Но у KNN при применении тратится больше ресурсов, чем у RF.

```python
y_pred = rf.predict(X_test)
print(classification_report(y_pred, y_test))
```
